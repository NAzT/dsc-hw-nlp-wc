```python
from pythainlp.tokenize import word_tokenize

def isThai(chr):
  cVal = ord(chr)
  if(cVal >= 3584 and cVal <= 3711):
    return True
  return False

tweets = pd.read_csv('shop.csv',encoding="utf8", header=None)
#comment = 'dasdกหก'


pureThai=[]
for comment in tweets[0]:
  pureThai.append(''.join(list(filter(isThai, comment))))

pureThai
text = pureThai
text = ' '.join(pureThai) # รวมแต่ละแถวเป็นบรรทัดเดียว
text = text.replace('ชิมช็อปใช้', '')
text = text.replace('เฟส', '')
# text
# text = text.replace("ลัลลาเบล", "")

wordcloud = WordCloud(font_path='THSarabunNew.ttf',
                      stopwords = ["และ", "หรือ", "ได้", "ให้", "ก็"] + list(thai_stopwords()),
                      background_color="white",
                      width=1024, # กว้าง
                      	height=768, # ยาว
                      	collocations=False,
                      regexp=r"[\u0E00-\u0E7Fa-zA-Z']+"
                      ).generate(' '.join(word_tokenize(text)))
#plt.imshow(wordcloud, cmap=plt.cm.gray, interpolation='bilinear')
#plt.axis("off")
#plt.show()
```

```python
# plt.figure(figsize=(15,4))
#fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
import matplotlib.pyplot as plt

width_in_inches = 15
height_in_inches = 15
dots_per_inch = 70

plt.figure(
    figsize=(width_in_inches, height_in_inches),
    dpi=dots_per_inch)

plt.imshow(wordcloud, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show() 
```
